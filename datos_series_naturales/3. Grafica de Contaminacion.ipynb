```python
import pandas
import numpy as np
import matplotlib.pyplot as plt

#%matplotlib notebook

def plot_signal_with_anomalies(file, title, start=None, end=None):
    dataframe = pandas.read_csv(file)
    if start is None:
        start = 0
    if end is None:
        end = len(dataframe)
    fig, axes = plt.subplots(constrained_layout=True, figsize=(12, 2.5))
    signal = dataframe['data'][start:end]
    labels = dataframe['label'][start:end]
    anomaly_indexes = np.where(np.asarray(labels) == 1)[0]
    axes.set_title(title)
    axes.set_xlim(start,end)
    axes.xaxis.set_ticks(np.arange(start, end + 1, 10000))
    axes.plot(signal, '', color='orange')
    axes.plot(np.take(signal, anomaly_indexes), '.', color='blue')
    axes.set_ylim(-1.1, 1.1)
    axes.set_xlabel('Tiempo')
    axes.set_ylabel('Señal')
    axes.grid(True)
    plt.show()


first = 0 
last = first + 100000
plot_signal_with_anomalies('grillos_y_ballena_v2_1porciento.csv', 'Grillos-Ballena 1%', first, last)    
plot_signal_with_anomalies('grillos_y_ballena_v2_2porciento.csv', 'Grillos-Ballena 2%', first, last)    
plot_signal_with_anomalies('grillos_y_ballena_v2_3porciento.csv', 'Grillos-Ballena 3%', first, last)    
plot_signal_with_anomalies('grillos_y_ballena_v2_5porciento.csv', 'Grillos-Ballena 5%', first, last)    
plot_signal_with_anomalies('grillos_y_ballena_v2_7porciento.csv', 'Grillos-Ballena 7%', first, last)    
plot_signal_with_anomalies('grillos_y_ballena_v2_10porciento.csv', 'Grillos-Ballena 10%',first, last)


```
